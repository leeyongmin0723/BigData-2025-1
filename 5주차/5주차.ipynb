```python
import pandas as pd
import requests
from folium.plugins import MarkerCluster
import folium
```

```python
url = 'http://openapi.seoul.go.kr:8088/476f4d426474726538325a7242556d/json/SeoulPublicLibraryInfo/1/100/'

response = requests.get(url)
```

```python
print(response.content)
```

```python
data = response.json()
data
```

```python
temp=[]
for d in data['SeoulPublicLibraryInfo']['row']:
  temp.append(d)
uni = pd.DataFrame(temp)
uni
```

```python
map = folium.Map(location=[37.4965, 126.9573],
                 zoom_start=11,
                 zoom_control=False,
                 control_scale=True)
map
```

```python
mc = MarkerCluster().add_to(map)

for n , row in uni.iterrows():
  folium.Marker(location=[row['XCNTS'], row['YDNTS']],
                popup = folium.Popup(row['LBRRY_NAME'], max_width=200)
                ).add_to(mc)

map
```

```python
map.save('5 .html')
```
