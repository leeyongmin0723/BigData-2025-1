```python
score2 = pd.read_excel('성적2.xlsx')
score2
```

```python
score2.info
```

```python
score2.dropna(subset=['출석'])
```

```python
score2.dropna(subset=['프로젝트','실험'])
```

```python
score2.dropna(axis=0) # axis=0 : 행삭제, axis=1 : 열 삭제 
```

```python
score2.dropna(axis=0,how='any') # any : 하나라도 포함하면 행/열 삭제, all : 모두 포함하면 행/열 삭제
```

```python
score2.dropna(axis=0,how='all') # any : 하나라도 포함하면 행/열 삭제, all : 모두 포함하면 행/열 삭제
```

```python
score2.dropna(axis=0, thresh =4, inplace=True) # 결측치가 4개 이상인것만 남기고 삭제
score2
```

```python
score3 = pd.read_excel('성적3.xlsx')
score3
```

```python
score3.interpolate() # 결측치 위 아래 값의 평균으로 대체
```

```python
score3.fillna(method = 'ffill') # 바로 앞이랑 똑같은 값으로 대체
```

```python
score3.fillna(method = 'bfill') # 바로 뒤랑 똑같은 값으로 대체
```

```python
score3.fillna(50) # 내가 원하는 값으로 지정해서 대체
```

```python
score3.fillna(score3['출석'].mean()) # 출석에 대한 평균으로 대체
```

```python
score3.fillna({'출석':10, '프로제트':20}) # 출석은 10, 프로젝트는 20으로 대체
```

```python
score4 = pd.read_excel('성적4.xlsx')
score4
```

```python
score4['성별'].value_counts().sort_index() #몇갠지 정리, 인덱스 정렬
```

```python
score4['성별'] = score4['성별'].replace(3, 2) # 3인 값을 2로 바꾸기
score4
```

```python
score4.duplicated(['ID']) #중복 되는차례 마다 True
```

```python
score4.drop_duplicates(['ID','성별']) #기본은 첫번째 데이터를 유지하고 뒤에 데이터 삭제
```

```python
score4.drop_duplicates(['ID','성별'], keep='last') # 먼저 데이터를 삭제하고 마지막 데이터를 유지
```

```python
score4.drop_duplicates(['ID','성별'], keep=False) #데이터 전부 삭제
```
